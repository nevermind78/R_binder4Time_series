# Simulate AR(p) model with R


$$ Y_t=a_1Y_{t-1}+a_2Y_{t-2}+\epsilon_t$$

```r
# noise
```

```r
e=rnorm(1500)
```

```r
a=c(-1.6,-0.8)
```

```r
y=filter(e,filter=a,method="recursive")
```

```r
ts.plot(y)
```

```r
y=arima.sim(model=list(ar=c(-1.6,-0.8)),n=250)
```

```r
ts.plot(y)
```

# MA(q)

$$ X_t=a_1\epsilon_{t-1}+a_2\epsilon_{t-2}+\epsilon_t$$

```r
a=c(0.4,0.5)
```

```r
y=arima.sim(model=list(ar=a),n=1000)
```

```r
ts.plot(y)
```
